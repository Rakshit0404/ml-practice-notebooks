2x - 3y = 15

4x + 10y = 14

draw plot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000)
```

```python
y1 = (15 - 2*x) / -3
y2 = (14 - 4*x) / 10
```

```python
fig, ax = plt.subplots()
plt.title("linear equation plot")
plt.xlabel("x")
plt.ylabel("y")
ax.set_xlim([0, 8])
ax.set_ylim([-5, 5])
ax.plot(x, y1, c = "green")
ax.plot(x, y2, c = "red")
plt.axvline(x = 6, color = "purple", linestyle = "--")
plt.axhline(y = -1, color = "purple", linestyle = "--")
```

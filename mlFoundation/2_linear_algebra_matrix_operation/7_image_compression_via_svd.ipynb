```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = Image.open("Vaman_avatar.jpg")
plt.imshow(img)
```

```python
img_gray = img.convert("LA")
plt.imshow(img_gray)
```

```python
img_mat = np.array(list(img_gray.getdata(band = 0)), dtype = float)
print(img_mat)
img_mat.shape = (img_gray.size[1], img_gray.size[0])
print(img_mat)
img_mat = np.matrix(img_mat)#just a precaultion
```

```python
U, d, VT = np.linalg.svd(img_mat)
```

```python
for i in [2, 4, 8, 16, 32, 64, 128, 256, 512]:
    reconstructed = np.dot(U[:, :i], np.dot(np.diag(d[:i]), VT[:i, :]))
    plt.imshow(reconstructed, cmap="gray")
    title = "n = %s" % i
    plt.title(title)
    plt.show()
```

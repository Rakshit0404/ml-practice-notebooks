```python
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
x = [0, 1, 2, 3, 4, 5, 6, 7.] # E.g.: Dosage of drug for treating Alzheimer's disease
y = [1.86, 1.31, .62, .33, .09, -.67, -1.23, -1.37] # E.g.: Patient's "forgetfulness score"
```

```python
fig, ax = plt.subplots()
plt.title("drug for treating Alzheimer's disease")
plt.xlabel("Dosage")
plt.ylabel("Forgetfullness")
ax.scatter(x, y)
```

```python
x1s = np.ones(8)
```

```python
x = np.concatenate(([x1s], [x]), axis = 0).T
x
```

```python
x_pinv = np.linalg.pinv(x)
x_pinv
```

```python
w = np.matmul(x_pinv, y)
w
```

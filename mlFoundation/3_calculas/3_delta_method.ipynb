```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000)
```

```python
y = x**2 + 2*x + 2
```

```python
x1 = 2
```

```python
y1 = x1**2 + 2*x1 + 2
```

```python
x2 = 5
```

```python
y2 = x2**2 + 2*x2 + 2
```

```python
m = (37-10)/(5-2)
```

```python
b = 37 - m*5
```

```python
line_y = m*x + b
```

```python
fig, ax = plt.subplots()
plt.title("graph for tangent")
plt.xlabel("X")
plt.ylabel("Y")
ax.set_xlim(-12, 12)
ax.set_ylim(-20, 150)
ax.axvline(x = 0, color = "lightgray")
ax.axhline(y = 0, color = "lightgray")
ax.plot(x, y)
ax.scatter(x1, y1, color = "orange", zorder = 3)
ax.scatter(x2, y2, color = "darkgreen", zorder = 3)
ax.plot(x, line_y, color = "tomato")
```

We can still see that the tangent doesnt look perfect. As close as we bring these two points to each other on the curve, the tangent keeps getting perfect.

```python
x1 = 2.1
x2 = 2.2
y1 = x1**2 + 2*x1 + 2
y2 = x2**2 + 2*x2 + 2
m = (y2 - y1)/(x2 - x1)
b = y2 - m*x2
y_line = m*x + b
```

```python
fig, ax = plt.subplots()
plt.title("graph for tangent")
plt.xlabel("X")
plt.ylabel("Y")
ax.set_xlim(-12, 12)
ax.set_ylim(-20, 150)
ax.axvline(x = 0, color = "lightgray")
ax.axhline(y = 0, color = "lightgray")
ax.plot(x, y)
ax.scatter(x1, y1, color = "orange", zorder = 3)
ax.scatter(x2, y2, color = "darkgreen", zorder = 3)
ax.plot(x, y_line, color = "tomato")
```

Now this looks perfect

```python
x = -1
deltaX = 0.00001
```

```python
y1 = x**2 + 2*x + 2
y1
```

```python
y2 = (x+deltaX)**2 + 2*(x+deltaX) + 2
y2
```

```python
m = (y2-y1)/deltaX
m
```

```python
b = y2 - m*(x+deltaX)
b
```

```python
x_line = np.linspace(-10, 10, 1000)
```

```python
y_line = m*x_line + b
```

```python
y_curve = x_line**2 + 2*x_line + 2
```

```python
fig, ax = plt.subplots()
plt.title("tangent for x = -1")
plt.xlabel("X")
plt.ylabel("Y")
ax.scatter(x, y1, color = "black")
ax.scatter(x, y2, color = "black", zorder = 3)
ax.plot(x_line, y_curve)
ax.plot(x_line, y_line)
ax.axvline(x = 0, color = "lightgray")
ax.axhline(y = 0, color = "lightgray")
```

```python
def calc_y(my_x):
    return my_x**2 + 2*my_x + 2
```

```python
def slope(calc_y, my_x, deltaX):
    dy = (calc_y(my_x + deltaX) - calc_y(my_x))/deltaX
    return dy
```

```python
deltas = [1, 0.1, 0.01, 0.001, 0.0001, 0.00001]
```

```python
for delts in deltas:
    print(slope(calc_y, 2, delts))
```

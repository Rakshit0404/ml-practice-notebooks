```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calc_y(my_x):
    y = (my_x**2 - 1)/(my_x - 1)
    return y
```

```python
calc_y(0.9)
```

```python
calc_y(0.999)
```

```python
calc_y(1.1)
```

```python
calc_y(1.001)
```

```python
def calc_y2(my_x):
    y = np.sin(my_x)/my_x
    return y
```

```python
calc_y2(0)
```

```python
calc_y2(0.001)
```

```python
calc_y2(-0.001)
```

```python
x = np.linspace(-10, 10, 1000)
y = np.sin(x)/x
```

```python
fig, ax = plt.subplots()
plt.title("graph for sin(x)/x")
plt.xlabel("X")
plt.ylabel("Y")
ax.set_xlim(-10, 10)
ax.set_ylim(-1, 2)
ax.axvline(x = 0, color = "purple", linestyle = "--")
ax.axhline(y = 1, color = "purple", linestyle = "--")
ax.plot(x, y)
```

LIMITS APPROACHING INFINITY

```python
def calc_y3(my_x):
    y = 25/my_x
    return y
```

```python
x = np.linspace(-10000, 10000, 100000)
y = calc_y3(x)
```

```python
fig, ax = plt.subplots()
plt.title("graph for 25/x")
plt.xlabel("X")
plt.ylabel("Y")
ax.set_xlim(-10, 10)
ax.set_ylim(-100, 100)
ax.plot(x, y)
```

We can see aquirk where matplot lib tries to connect the plot from -infi to +infi on y axis. So we split the array

```python
x1 = x[x<0]
x2 = x[x>0]
```

```python
y1 = calc_y3(x1)
y2 = calc_y3(x2)
```

```python
fig, ax = plt.subplots()
plt.title("graph for 25/x")
plt.xlabel("X")
plt.ylabel("Y")
ax.set_xlim(-10, 10)
ax.set_ylim(-250, 250)
ax.axvline(x = 0, color = "lightgray")
ax.axhline(y = 0, color = "lightgray")
ax.plot(x1, y1)
ax.plot(x2, y2)
plt.show()
```

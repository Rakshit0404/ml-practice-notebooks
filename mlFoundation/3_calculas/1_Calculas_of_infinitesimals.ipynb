```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000)
```

```python
y = x**2 + 2*x + 2
```

```python
fig, ax = plt.subplots()
plt.title("Learning infinitesimals")
plt.xlabel("X")
plt.ylabel("Y")
ax.set_xlim(-12, 12)
ax.set_ylim(0, 150)
ax.plot(x, y, color = "green")
```

```python
fig, ax = plt.subplots()
plt.title("Learning infinitesimals")
plt.xlabel("X")
plt.ylabel("Y")
ax.set_xlim(-1.01, -0.99)
ax.set_ylim(0.99, 1.01)
ax.plot(x, y, color = "green")
```

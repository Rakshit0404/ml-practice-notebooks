```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import math
```

```python
def getZ(my_x, my_y):
    return my_x**2 - my_y**2
```

```python
x = torch.linspace(-3, 3, 1000, requires_grad=True)
```

```python
z = getZ(x, 0)
```

```python
fig, ax = plt.subplots()
plt.xlabel("X")
plt.ylabel("Z")
plt.title("relation between x and z")
ax.axvline(x = 0, color = "lightgray")
ax.axhline(y = 0, color = "lightgray")
ax.plot(x.detach().numpy(), z.detach().numpy())
```

```python
x_values = torch.tensor([-2, -1, 1, 2.]).requires_grad_(True)
x_values
```

```python
fig, ax = plt.subplots()
plt.xlabel("X")
plt.ylabel("Z")
plt.title("relation between x and z")
ax.set_ylim(-2, 10)
ax.set_xlim(-3.5, 3.5)
ax.axvline(x = 0, color = "lightgray")
ax.axhline(y = 0, color = "lightgray")
ax.plot(x.detach().numpy(), z.detach().numpy())
for x_val in x_values:
    my_x = torch.tensor(x_val, requires_grad=True)
    print(my_x)
    my_z = my_x**2
    my_z.backward()
    my_m = my_x.grad
    print(my_m)
    b= my_z-my_m*my_x
    z = my_m*x +b # As we know b = 0
    ax.plot(x.detach().numpy(), z.detach().numpy())
    ax.scatter(my_x.detach().numpy(), my_z.detach().numpy(), zorder = 3)

plt.show()
```
